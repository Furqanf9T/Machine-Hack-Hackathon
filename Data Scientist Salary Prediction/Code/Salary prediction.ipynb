```python
def conversion(tmp):

    coldif=[]
    for i in range(0,len(tmp)):
        for j in range(0,len(tmp[i])):
            coldif.append(tmp[i][j+1]-tmp[i][j])
            break

    coldif=np.array(coldif)/2

    colsum=[]
    for i in range(0,len(tmp)):
        for j in range(0,len(tmp[i])):
            colsum.append(int(np.mean((tmp[i][j+1]-coldif[i],tmp[i][j]+coldif[i]))))
            break
    tmp=colsum
    return tmp
```
